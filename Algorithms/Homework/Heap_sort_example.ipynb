```python
from random import randint
```

```python
def build_student_list(n):
    students = []
    for i in range(1,n):
        grad = randint(0,1)
        if grad == 1:
            yr = 5
        else:
            yr = randint(0,4)
        if yr != 5 and yr != 0:
            major = randint(0,1)
        else:
            major = 0
        reg = randint(0,5)
        
        students.append(('student' + str(i), int(str(grad)+str(yr)+str(major)+str(reg))))
    return students
```

```python
students = build_student_list(100)
```

```python
def max_heapify(A, stop, n = 0, i = 0):
    '''
    INPUT:  (A)    - a list,
            (stop) - an index position indicating how much of the list to max_heapify,
            (n)    - a passthrough step counting variable,
            (i)    - an index position indicating where to begin to heapify
    RETURN: (n)    - a step count
    '''
    largest = i
    left_arg = 2*i + 1
    right_arg = 2*i + 2
    n+=4                                                        # assign variables and check if
    if left_arg < stop:
        n+=1                                                    # check if
        if left_arg < (stop) and A[left_arg][1] > A[i][1]:
            n+=1                                                # assign largest
            largest = left_arg
        else:
            n+=1                                                # assign largest
            largest = i 
    
    n+=1                                                        # check if
    if right_arg < stop:
        n+=1                                                    # check if
        if right_arg < (stop) and A[right_arg][1] > A[largest][1]:
            n+=1.                                               # assign largest
            largest = right_arg
    n+=1                                                        # check if
    if largest != i:
        n+=1                                                    # make switch
        A[i], A[largest] = A[largest], A[i]
        max_heapify(A,stop = stop,n=n, i=largest)

    return n
```

```python
def build_max_heap(A, n=0):
    '''
    INPUT:  (A)    - a list,
            (n)    - a passthrough step counting variable
    RETURN: (n)    - a step count
    '''
    x = len(A)
    n+= 1                                           # assign x
    for i in range(int(x/2), -1, -1):
        n+=1                                        # assign i
        max_heapify(A,stop=len(A), n=n ,i=i)
    return n
```

```python
def heap_sort(A, n = 0):
    '''
    INPUT:  (A)    - a list,
            (n)    - a passthrough step counting variable
    RETURN: (n)    - a step count
    '''
    n += build_max_heap(A)
    for i in range((len(A)-1),1, -1):
        n+= 2                                      # assign i and make switch
        A[0], A[i] = A[i], A[0]
        max_heapify(A, stop=(i-1), n=n)
    return n
```
