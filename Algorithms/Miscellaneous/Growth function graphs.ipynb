```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import math
%matplotlib inline
```

```python
def iterativeLog(n, i=1):
    from numpy import log2
    if log2(n) <= 1:
        return i
    else:
        i += 1
        return iterativeLog(log2(n),i)
```

```python
X = np.linspace(1,100,100)
y1 = X**2
y2 = X**3
y3 = 2**X
y4 = [np.log2(iterativeLog(x)) for x in X]
y5 = np.log(X)
y6 = np.ones(100)
y7 = X
y8 = np.e**X
y9 = [np.log(np.log(x)) for x in X if x != 0]
y10 = (3/2)**X
y11 = 2**(np.log2(X))
y12 = [2**(iterativeLog(x)) for x in X]
y13 = (np.sqrt(2))**(np.log2(X))
y14 = [math.factorial(x) for x in X]
y15 = [math.factorial(int(np.log2(x))) for x in X]
y16 = np.log2(X)**2
y17 = [np.log2(x) for x in [math.factorial(x) for x in X][:20]]
y18 = 2**(2**X)
y19 = [x**(1/np.log2(x)) for x in X]
y20 = X*2**X
y21 = [x**(np.log2(np.log2(x))) for x in X]
y22 = [(np.log2(x))**(np.log2(x)) for x in X]
y23 = [4**(np.log2(x)) for x in X]
y24 = [math.factorial(x+1) for x in X]
y25 = [np.sqrt(np.log2(x)) for x in X]
y26 = [2**(np.sqrt(2*np.log2(x))) for x in X]
y27 = [x*np.log2(x) for x in X]
y28 = [2**2**(x+1) for x in X]


plt.figure(figsize=(20,12))
plt.ylim(0,70)
plt.xlim(0,50)
plt.plot(X,y1,'b', label= '$n^2$')
plt.plot(X,y2,'g', label= '$n^3$')
plt.plot(X,y3,'orange', label= '$2^n$')
plt.plot(X,y4,'cyan', label= '$lg^*(n)$')
plt.plot(X,y5, 'magenta', label= '$ln(n)$')
plt.plot(X,y6, 'purple', label= '$1$')
plt.plot(X,y7,'b', linestyle = '--', label= '$n$')
plt.plot(X,y8,'g', linestyle = '--', label= '$e^n$')
plt.plot(X,y9,'orange', linestyle = '--', label= '$lg(lg(n))$')
plt.plot(X,y10,'cyan', linestyle = '--', label= '$(3/2)^n$')
plt.plot(X,y11, 'magenta', linestyle = '--', label= '$2^{lg(n)}$')
plt.plot(X,y12, 'purple', linestyle = '--', label= '$2^{lg^*(n)}$')
plt.plot(X,y13,'b', linestyle = '-.', label= '$\sqrt{2}**{lg(n)}$')
plt.plot(X,y14,'g', linestyle = '-.', label= '$n!$')
plt.plot(X,y15,'orange', linestyle = '-.', label= '$(lg(n))!$')
plt.plot(X,y16,'cyan', linestyle = '-.', label= '$lg(n)^2$')
plt.plot(X[:20],y17, 'magenta', linestyle = '-.', label= '$lg(n!)$')
plt.plot(X,y18, 'purple', linestyle = '-.', label= '$2^{2^n}$')
plt.plot(X,y19,'b', linestyle = ':', label= '$n^{1/lg(n)}$')
plt.plot(X,y20,'g', linestyle = ':', label= '$n*2^n$')
plt.plot(X,y21,'orange', linestyle = ':', label= '$n^{lg(lg(n))}$')
plt.plot(X,y22,'cyan', linestyle = ':', label= '$lg(n)^{lg(n)}$')
plt.plot(X,y23, 'magenta', linestyle = ':', label= '$4^{lg(n)}$')
plt.plot(X,y24, 'purple', linestyle = ':', label= '$(n+1)!$')
plt.plot(X,y25,'orange', marker = 'o', label= '$\sqrt{lg(n)}$')
plt.plot(X,y26,'cyan', marker = 'o', label= '$2^{\sqrt{2*lg(n)}}$')
plt.plot(X,y27, 'magenta', marker = 'o', label= '$n*lg(n)$')
plt.plot(X,y28, 'purple', marker='o', label = '$2^{2^{n+1}}$')




plt.legend(fontsize = 'large',bbox_to_anchor = (-0.03,1))

plt.savefig('growth_rates.png')
```

```python
labels = ['$n^2$','$n^3$','$2^n$','$lg^*(n)$','$ln(n)$','$1$','$n$','$e^n$','$lg(lg(n))$','$(3/2)^n$','$2^{lg(n)}$',
         '$2^{lg^*(n)}$','$\sqrt{2}**{lg(n)}$','$n!$','$(lg(n))!$','$lg(n)^2$','$lg(n!)$','$2^{2^n}$','$n^{1/lg(n)}$',
         '$n*2^n$','$n^{lg(lg(n))}$','$lg(n)^{lg(n)}$','$4^{lg(n)}$','$(n+1)!$','$\sqrt{lg(n)}$','$2^{\sqrt{2*lg(n)}}$',
         '$n*lg(n)$','$2^{2^{n+1}}$']
data = [y1, y2, y3, y4, y5, y6, y7, y8, y9, y10, y11, y12, y13, y14, y15, y16, y17, y18, y19, y20,
       y21, y22, y23, y24, y25, y26, y27, y28]
```

```python
df = pd.DataFrame(data, index = labels, columns = [x for x in range(1, 101)])
```
