```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import 
%matplotlib inline
```

```python
covid_sum = pd.read_csv('covidSummary.csv', encoding = "ISO-8859-1")
```

```python
covid_sum.head()
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(covid_sum.isna())
plt.tight_layout()
plt.savefig('na_heatmap.png')
```

```python
for c in list(covid_sum.columns):
    col = covid_sum.loc[:,c]
    na = len(col.dropna())
    l = len(col)
    if na/l < 0.25:
        covid_sum = covid_sum.drop(c,axis = 1)
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(covid_sum.isna())
plt.tight_layout()
plt.savefig('na_heatmap.png')
```
